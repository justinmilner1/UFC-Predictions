```python
df = pd.read_csv('../../data/JMpreprocessed_UnivariateSelecton_<5.csv')
```

```python
#splitting up data into training/testing
y = df['Winner']
X = df.drop(columns = 'Winner')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=43)
```

```python
'''
This function trains/test and returns the accuracy of that iteration
'''
def get_score(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    return model.score(x_test, y_test)
```

```python
best_score = 0
best_depth = 0
best_subsample = 0
best_learning_rate = 0
best_early_stop = 0
for depth in [4, 10, 16, 22]:
    print("depth: " + str(depth))
    for subsample in [.5, 1]:
        print("subsample: " + str(subsample))
        for estimator in [50, 100, 250, 400]:
            for learning_rate in [.001, .01, .1]:
                for early_stop in [10, 25]:
                    xgboost_model = xgb.XGBClassifier(max_depth=depth,
                                            subsample=subsample,
                                            objective='binary:logistic',
                                            n_estimators=estimator,
                                            learning_rate = learning_rate,
                                            early_stopping_rounds=early_stop)
                    score = get_score(xgboost_model, X_train, X_test, y_train, y_test)

                    if score > best_score:
                        best_score = score
                        best_depth = depth
                        best_subsample = subsample
                        best_learning_rate = learning_rate
                        best_early_stop = early_stop
                    

                    
                
```

```python
print("Best Score: " + str(best_score)+ '\n' +
     "Best Depth: " + str(best_depth) + '\n' +
     "Best Subsample: " + str(best_subsample) + '\n' +
      "Best learning rate: " + str(best_learning_rate) + '\n'
     )
```
